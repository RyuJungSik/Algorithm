```python
n=int(input())
arr=[]
for _ in range(n):
    a,b=map(int,input().split())
    arr.append((a,b))
ans=0 
def dfs(index, working, price):
    global ans
    if index==n:
        ans=max(ans, price)
        return
    else:
        if working!=0:
            working-=1
        if working==0 and index+arr[index][0]<n+1:
            dfs(index+1,arr[index][0],price+arr[index][1])
        dfs(index+1,working,price)
            

dfs(0,0,0)
print(ans)
```
