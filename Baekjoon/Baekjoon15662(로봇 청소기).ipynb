```python
n,m=map(int, input().split())


#0북 1동 2남 3서
x,y,d=map(int, input().split())

#0빈칸 1벽
board=[]
for _ in range(n):
    board.append(list(map(int, input().split())))

dx=[0,-1,0,1]
dy=[-1,0,1,0]
bx=[1,0,-1,0]
by=[0,-1,0,1]
check=[[False]*m for _ in range(n)]    
rotate=0
while True:
    check[x][y]=True
    
    if rotate==4:
        nx=x+bx[d]
        ny=y+by[d]

        if 0<=nx<n and 0<=ny<m and board[nx][ny]==0:
            check[nx][ny]=True
            x,y=nx,ny
            rotate=0
            continue
        else:
            break
    else:
        nx=x+dx[d]
        ny=y+dy[d]
  
        if 0<=nx<n and 0<=ny<m and board[nx][ny]==0 and check[nx][ny]==False:
            check[nx][ny]=True
            x=nx
            y=ny
            rotate=0
            if d==0:
                d=3
            else:
                d-=1
            continue
        else:
            rotate+=1
            if d==0:
                d=3
            else:
                d-=1
            continue
answer=0

for i in check:
    for j in i:
        if i==True:
            answer+=1
print(answer)
```
