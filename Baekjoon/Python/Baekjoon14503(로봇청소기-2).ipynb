```python
def rotate(d):
    if d==0:
        return 3
    if d==1:
        return 0
    if d==2:
        return 1
    if d==3:
        return 2

def nextPos(i,j,d):
    if d==0:
        return (i-1,j)
    if d==1:
        return (i,j+1)
    if d==2:
        return (i+1,j)
    if d==3:
        return (i,j-1)
    
def nextBack(i,j,d):
    if d==0:
        return (i+1,j)
    if d==1:
        return (i,j-1)
    if d==2:
        return (i-1,j)
    if d==3:
        return (i,j+1)
n,m=map(int,input().split())
x,y,d=map(int,input().split())  #0 북  1 동  2 남  3 서
board=[]
answer=1
for i in range(n):
    board.append(list(map(int,input().split())))

while True:
    board[x][y]=-1
    nd=d
    flag=False
    for i in range(4):
        nd=rotate(nd)
        (nx,ny)=nextPos(x,y,nd)
        if board[nx][ny]==0:
            x,y=nx,ny
            d=nd
            flag=True
            break
    if flag:
        if board[x][y]==0:
            answer+=1
        continue
    else:
        (nx,ny)=nextBack(x,y,d)
        if board[nx][ny]==1:
            break
        else:
            x,y=nx,ny
            if board[x][y]==0:
                answer+=1
print(answer)
```
