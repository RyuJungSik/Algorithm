```python
from collections import deque

def bfs(x,y):
    global tmp,visited,n,l,r,board,s
    q=deque([(x,y)])
    visited[x][y]=True
    dx,dy=[0,0,1,-1],[1,-1,0,0]
    while q:
        nowX,nowY=q.popleft()
        tmp.append((nowX,nowY))
        s+=board[nowX][nowY]
        for d in range(4):
            nx,ny=nowX+dx[d],nowY+dy[d]
            if nx<0 or nx>=n or ny<0 or ny>=n or visited[nx][ny]==True:
                continue
            if l<=abs(board[nowX][nowY]-board[nx][ny])<=r:
                visited[nx][ny]=True
                q.append((nx,ny))
            


n,l,r=map(int,input().split())
board=[]
for _ in range(n):
    board.append(list(map(int,input().split())))
flag=True
cnt=0
while True:
    cnt+=1
    visited=[[False]*n for _ in range(n)]
    flag=False
    for i in range(n):
        for j in range(n):
            if visited[i][j]==True:
                continue
            tmp=[]
            s=0
            visited[i][j]=True
            bfs(i,j)
            if len(tmp)>1:
                flag=True
            for t in tmp:
                x,y=t[0],t[1]
                board[x][y]=s//len(tmp)
    
    
    if flag==False:
        break
print(cnt-1)
```
