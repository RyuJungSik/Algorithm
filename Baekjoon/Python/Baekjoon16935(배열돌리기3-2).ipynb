```python
import copy


def oper1(board,n,m):
    for j in range(m):
        tmp=[]
        for i in range(n):
            tmp.append(board[i][j])
        cnt=0
        for i in range(n):
            board[n-i-1][j]=tmp[cnt]
            cnt+=1
    return board

def oper2(board,n,m):
    for i in range(n):
        board[i]=list(reversed(board[i]))
    return board
    
def oper3(board,n,m):
    boardCopy=[[0]*n for _ in range(m)]
    for i in range(n):
        tmp=[]
        for j in range(m):
            tmp.append(board[i][j])
        cnt=0
        for j in range(m):
            boardCopy[j][n-1-i]=tmp[cnt]
            cnt+=1
    return boardCopy
            
def oper4(board,n,m):
    boardCopy=[[0]*n for _ in range(m)]
    for i in range(n):
        tmp=[]
        for j in range(m):
            tmp.append(board[i][j])
        cnt=0
        for j in range(m):
            boardCopy[m-j-1][i]=tmp[cnt]
            cnt+=1
    return boardCopy


def oper5(board,n,m):
    boardCopy=copy.deepcopy(board)
    for i in range(n//2):
        for j in range(m//2):
            board[i][m//2+j]=boardCopy[i][j]
    for i in range(n//2):
        for j in range(m//2,m):
            board[n//2+i][j]=boardCopy[i][j]
    for i in range(n//2,n):
        for j in range(m//2,m):
            board[i][j-m//2]=boardCopy[i][j]
    for i in range(n//2,n):
        for j in range(m//2):
            board[i-(n//2)][j]=boardCopy[i][j]
    return board


def oper6(board,n,m):
    boardCopy=copy.deepcopy(board)
    for i in range(n//2):
        for j in range(m//2):
            board[i][m//2+j]=boardCopy[i][j]
    for i in range(n//2):
        for j in range(m//2,m):
            board[n//2+i][j]=boardCopy[i][j]
    for i in range(n//2,n):
        for j in range(m//2,m):
            board[i][j-m//2]=boardCopy[i][j]
    for i in range(n//2,n):
        for j in range(m//2):
            board[i-(n//2)][j]=boardCopy[i][j]
    return board


def oper6(board,n,m):
    boardCopy=copy.deepcopy(board)
    for i in range(n//2):
        for j in range(m//2):
            board[i+(n//2)][j]=boardCopy[i][j]
    for i in range(n//2):
        for j in range(m//2,m):
            board[i][j-(m//2)]=boardCopy[i][j]
    for i in range(n//2,n):
        for j in range(m//2,m):
            board[i-(n//2)][j]=boardCopy[i][j]
    for i in range(n//2,n):
        for j in range(m//2):
            board[i][j+(m//2)]=boardCopy[i][j]
    return board


n,m,r=map(int,input().split())
board=[]
for _ in range(n):
    board.append(list(map(int,input().split())))
oper=list(map(int,input().split()))

for i in oper:
    if i==1:
        board=oper1(board,n,m)
    elif i==2:
        board=oper2(board,n,m)
    elif i==3:
        board=oper3(board,n,m)
        n,m=m,n
    elif i==4:
        board=oper4(board,n,m)
        n,m=m,n
    elif i==5:
        board=oper5(board,n,m)
    elif i==6:
        board=oper6(board,n,m)
        
for i in board:
    print(" ".join(list(map(str,i))))
```
