```python
s,n=map(int, input().split())
xLen=2*s+3
yLen=s+2


def makeOne(arr):
    for i in range(xLen):
        if i==0:
            continue
        if i==(xLen//2):
            continue
        if i==xLen-1:
            continue
        arr[i][yLen-1]="|"


def makeTwo(arr):
    #가로 이걸로
    for i in [0,xLen//2,xLen-1]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    # 세로 이거이용
    for i in [yLen-1,0]:
        if i==yLen-1:
            for j in range(xLen//2):
                if j==0:
                    continue
                arr[j][i]="|"
        else:
             for j in range(xLen//2):
                if j==0:
                    continue
                arr[xLen//2+j][i]="|"
            
def makeThree(arr):
    for i in [0,xLen//2,xLen-1]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    
    for i in [yLen-1]:
        for j in range(xLen):
            if j==0 or j==xLen//2 or j==xLen-1:
                continue
            arr[j][i]="|"
            
def makeFour(arr):
    for i in [xLen//2]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    
    for i in [0, yLen-1]:
        if i==yLen-1:
            for j in range(xLen):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"
        else:
             for j in range(xLen//2):
                if j==0:
                    continue
                arr[j][i]="|"
                
def makeFive(arr):
    #가로 이걸로
    for i in [0,xLen//2,xLen-1]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    # 세로 이거이용
    for i in [0,yLen-1]:
        if i==0:
            for j in range(xLen//2):
                if j==0:
                    continue
                arr[j][i]="|"
        else:
             for j in range(xLen//2):
                if j==0:
                    continue
                arr[xLen//2+j][i]="|"

def makeSix(arr):
    #가로 이걸로
    for i in [0,xLen//2,xLen-1]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    # 세로 이거이용
    for i in [0,yLen-1]:
        if i==0:
            for j in range(xLen):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"
        else:
             for j in range(xLen//2):
                if j==0:
                    continue
                arr[xLen//2+j][i]="|"
        
def makeSeven(arr):
    #가로 이걸로
    for i in [0]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    # 세로 이거이용
    for i in [yLen-1]:
        for j in range(xLen):
            if j==0 or j==xLen//2 or j==xLen-1:
                continue
            arr[j][i]="|"
            
def makeEight(arr):
    #가로 이걸로
    for i in [0,xLen//2,xLen-1]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    # 세로 이거이용
    for i in [0,yLen-1]:
        if i==0:
            for j in range(xLen):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"
        else:
             for j in range(xLen):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"

def makeNine(arr):
    #가로 이걸로
    for i in [0,xLen//2,xLen-1]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    # 세로 이거이용
    for i in [0,yLen-1]:
        if i==0:
            for j in range(xLen//2):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"
        else:
             for j in range(xLen):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"
                
def makeZero(arr):
    #가로 이걸로
    for i in [0, xLen-1]:
        for j in range(yLen):
            if j==0 or j==yLen-1:
                continue
            arr[i][j]="-"
    # 세로 이거이용
    for i in [0,yLen-1]:
        if i==0:
            for j in range(xLen):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"
        else:
             for j in range(xLen):
                if j==0 or j==xLen//2 or j==xLen-1:
                    continue
                arr[j][i]="|"
                
allA=[[] for _ in range(xLen)]

for i in str(n):
    arr=[[" "]*yLen for _ in range(xLen)]  
    if i=='1':
        makeOne(arr)
    elif i=='2':
        makeTwo(arr)
    elif i=='3':
        makeThree(arr)
    elif i=='4':
        makeFour(arr)
    elif i=='5':
        makeFive(arr)
    elif i=='6':
        makeSix(arr)
    elif i=='7':
        makeSeven(arr)
    elif i=='8':
        makeEight(arr)
    elif i=='9':
        makeNine(arr)
    elif i=='0':
        makeZero(arr)
    
    for j in range(xLen):
        allA[j]=allA[j]+arr[j]+[" "]
        
for i in allA:
    print("".join(i))
```
