```python
from collections import deque

def nextStep(x,y,d):
    if d==0:
        return (x,y+1)
    elif d==1:
        return (x+1,y)
    elif d==2:
        return (x,y-1)
    else:
        return (x-1,y)

def makeDis(now,change):
    if change=="X":
        return now
    if now==0:
        if change=="L":
            return 3
        if change=="D":
            return 1
    elif now==1:
        if change=="L":
            return 0
        else:
            return 2
    elif now==2:
        if change=="L":
            return 1
        else:
            return 3
    else:
        if change=="L":
            return 2
        else:
            return 0
n=int(input())
k=int(input())
board=[[0]*n for _ in range(n)]
order=["X"]*10001
for _ in range(k):
    x,y=map(int,input().split())
    board[x-1][y-1]=1
    
l=int(input())
for _ in range(l):
    s=input()
    x,c=s.split()
    order[int(x)]=c
    
# 0 동쪽  1 남쪽  2  서쪽  3 북쪽
answer=0
dis=0
board[0][0]=2
q=deque([])
q.appendleft((0,0))
while True:
    answer+=1
    x,y=q[0]
    (nx,ny)=nextStep(x,y,dis)
    if nx<0 or nx>=n or ny<0 or ny>=n or board[nx][ny]==2:
        print(answer)
        break
    if board[nx][ny]==1:
        q.appendleft((nx,ny))
        board[nx][ny]=2
    else:
        q.appendleft((nx,ny))
        board[nx][ny]=2
        lx,ly=q.pop()
        board[lx][ly]=0
    dis=makeDis(dis,order[answer])
```
