```python
from collections import deque
from itertools import permutations
import copy

def solution(board, r, c):
    answer = []
    hash={}
    def bfs(board,sx,sy,ex,ey):
        visited=[[False]*4 for _ in range(4)]
        q=deque([(sx,sy,0)])
        visited[sx][sy]=True
        dx,dy=[0,0,1,-1],[1,-1,0,0]
        while q:
            nowX,nowY,nowClick=q.popleft()
            if nowX==ex and nowY==ey:
                return nowClick
            for d in range(4):
                nx=nowX+dx[d]
                ny=nowY+dy[d]
                if nx<0 or nx>=4 or ny<0 or ny>=4:
                    continue
                if visited[nx][ny]: continue
                visited[nx][ny]=True
                q.append([nx,ny,nowClick+1])
                
            for d in range(4):
                nx=nowX+dx[d]
                ny=nowY+dy[d]
                while True:
                    if nx<0 or nx>=4 or ny<0 or ny>=4:
                        nx-=dx[d]
                        ny-=dy[d]
                        break
                    if board[nx][ny]!=0:
                        break
                    nx+=dx[d]
                    ny+=dy[d]
                if visited[nx][ny]: continue
                visited[nx][ny]=True
                q.append([nx,ny,nowClick+1])
                
                
    
    def dfs(index,per,nowX,nowY,nowS):
        if index==len(per):
            answer.append(nowS)
            return 
        else:
            oneX,oneY=hash[per[index]][0]
            twoX,twoY=hash[per[index]][1]
            
            oneD1=bfs(board,nowX,nowY,oneX,oneY)
            oneD2=bfs(board,oneX,oneY,twoX,twoY)
            board[oneX][oneY]=0
            board[twoX][twoY]=0
            dfs(index+1,per,twoX,twoY,nowS+oneD1+oneD2+2)
            board[oneX][oneY]=per[index]
            board[twoX][twoY]=per[index]
            
            twoD1=bfs(board,nowX,nowY,twoX,twoY)
            twoD2=bfs(board,twoX,twoY,oneX,oneY)
            board[oneX][oneY]=0
            board[twoX][twoY]=0
            dfs(index+1,per,oneX,oneY,nowS+twoD1+twoD2+2)
            board[oneX][oneY]=per[index]
            board[twoX][twoY]=per[index]
            
        
        
    for i in range(4):
        for j in range(4):
            if board[i][j]!=0:
                if board[i][j] not in hash:
                    hash[board[i][j]]=[[i,j]]
                else:
                    hash[board[i][j]].append([i,j])
    
    for per in permutations(hash.keys(),len(hash.keys())):
        dfs(0,per,r,c,0)
    
    
    return min(answer)
```
